```python
import numpy as np

```

```python
with open("Loan_prediction_dataset.csv", "r") as file:
  data = np.genfromtxt(file ,delimiter=',', dtype="str")
```

```python
subset = data[1: ,8]
```

```python

subset[subset==""]= "0"
subset=subset.astype(int)
```

```python
mean_data = np.mean(subset)
median_data = np.median(subset)
std_deviation = np.std(subset)

print(f"Mean; {mean_data}")
print(f"Median: {median_data}")
print(f"Standard Deviation: {std_deviation}")
```
