```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd
```

```python
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
# Convert the toy dataset to a pandas DataFrame
df = pd.DataFrame(toy_dataset)
df
```

```python
df_encoded = pd.get_dummies(df.stack()).groupby(level=0).sum()
```

```python
data = df_encoded.applymap(lambda x: 1 if x != 0 else 0)
data
```

```python
# Run the Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(data, min_support=0.2, use_colnames=True)
```

```python
# Derive association rules from the frequent itemsets
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
```

```python
# Display the results
print(frequent_itemsets)
```

```python
print(rules)
```
