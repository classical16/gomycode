```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
for column in ['OsPlatformSubRelease', 'Census_OSEdition']:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
target_variable = ['HasDetections']
features = ['Census_IsVirtualDevice', 'Census_OSEdition', 'Census_HasOpticalDiskDrive']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[features],df[target_variable], test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#Scale numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# Handle class imbalance (for classification tasks)
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_scaled, y_train)
```

```python
# Convert the NumPy arrays to Pandas DataFrames
X_train_balanced_df = pd.DataFrame(X_train_balanced)
X_test_scaled_df = pd.DataFrame(X_test_scaled)
y_train_balanced_df = pd.DataFrame(y_train_balanced)
y_test_df = pd.DataFrame(y_test)
```

```python
# Save the prepared dataset for modeling
X_train_balanced_df.to_csv('X_train.csv', index=False)
X_test_scaled_df.to_csv('X_test.csv', index=False)
y_train_balanced_df.to_csv('y_train.csv', index=False)
y_test_df.to_csv('y_test.csv', index=False)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
```

```python
# Initialize the DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
# Fit the model using the training data
model.fit(X_train, y_train)
```

```python
# Predict probabilities for the positive class (1) on the test data
y_probs = model.predict_proba(X_test)[:, 1]
```

```python
# Calculate the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
```

```python
# Calculate the AUC score
auc_score = roc_auc_score(y_test, y_probs)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'Decision Tree (AUC = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
```

```python
# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Visualize the clusters
plt.scatter(df['Census_IsVirtualDevice'], df['Census_OSEdition'], c=df['cluster'], cmap='viridis')
plt.xlabel('Census_IsVirtualDevice')
plt.ylabel('Census_OSEdition')
plt.title('K-means Clustering')
plt.show()
```

```python
# Initialize an empty list to store the inertia values
inertia = []
```

```python
# Iterate over a range of K values
for k in range(1, 11,):
   # Apply K-means clustering
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    # Append the inertia value to the list
    inertia.append(kmeans.inertia_)
```

```python
# Plot the inertia values for different K values
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

Finding the optimal number of clusters (K) in K-means clustering can be achieved through several methods, such as the Elbow method and the Silhouette score.

Elbow Method:
The Elbow method is a graphical technique to identify the optimal number of clusters in a dataset. The idea is to plot the explained variance as a function of the number of clusters, and to look for the "elbow" point where the explained variance starts to decrease at a slower rate. This point indicates that adding more clusters does not significantly improve the explained variance, and is often a good choice for the optimal number of clusters.
