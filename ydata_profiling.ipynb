```python
import pandas as pd

from ydata_profiling import ProfileReport
```

```python
df = pd.read_csv("/content/EDUCATION_ATTAINMENT.csv")
```

```python
df.info
```

```python
# descriptive analysis
description = df.describe
print(description)
```

```python
# generate profile report
profile =ProfileReport(df,title='Education Attainment And Enrolment Around The World', explorative=True)

# display the report
profile.to_notebook_iframe()
```

```python
# generate an HTML report
profile.to_file("Education Attainment And Enrolment Around The World.html")
```
